```python
import os
import sys
sys.path.append('../')
```

```python
from pose_markup.video_processor import VideoProcessor
```

```python
INPUT_PATH = os.path.join(os.pardir, "data/")
OUTPUT_PATH = os.path.join(os.pardir, "output_data/")
MODEL_PATH = os.path.join(os.pardir, "model/pose_landmarker_full.task")

if not os.path.exists(OUTPUT_PATH):
    os.mkdir(OUTPUT_PATH)

video_proc = VideoProcessor(MODEL_PATH, OUTPUT_PATH)
```

```python
names = sorted(os.listdir(INPUT_PATH))
names[:10]
```

```python
for name in names:
    processed_time = video_proc.process_video(INPUT_PATH + name)
    print(f"Video {name} was processed for a {processed_time / 1000} s")
```

```python
not_processed_videos = []

for name in names[68:]:
    try:
        processed_time = video_proc.process_video(INPUT_PATH + name)
        print(f"Video {name} was processed for a {processed_time / 1000} s")
    except:
        not_processed_videos.append(name)
        print(f"Video {name} was not processed. ")
```
